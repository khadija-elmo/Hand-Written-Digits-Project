```python
import os
import numpy as np
import matplotlib.pyplot as plt
import cv2  #computer vision 
import tensorflow as tf
```

```python
#Load the MNIST data from TensoFlow 
mnist = tf.keras.datasets.mnist
```

```python
#Splitting the data into training and testing Data via load_data() that returns 2 tuples
(x_train,y_train),(x_test,y_test) = mnist.load_data()
```

```python
#normalizing the inputs data so to have values between 0 and 1 
x_train = tf.keras.utils.normalize(x_train, axis=1)
x_test = tf.keras.utils.normalize(x_test, axis=1)
```

```python
x_train.shape
```

```python
#Resising image to make it suitable for apply Convolution operation
img_size = 28
x_trainr = np.array(x_train).reshape(-1,img_size, img_size, 1)
x_testr = np.array(x_test).reshape(-1,img_size, img_size, 1)
print("training samples dimension : ",x_trainr.shape)
print("testing samples dimension : ",x_testr.shape)
```

```python
#Creating a model
model = tf.keras.models.Sequential()
```

```python
#Add layers to our model 

#first Convolutionnal layer
model.add(tf.keras.layers.Conv2D(64,(3,3),input_shape = x_trainr.shape[1:]))
model.add(tf.keras.layers.Activation("relu"))
model.add(tf.keras.layers.MaxPooling2D(pool_size=(2,2)))

#Secondt Convolutionnal layer
model.add(tf.keras.layers.Conv2D(64,(3,3),input_shape = (28,28)))
model.add(tf.keras.layers.Activation("relu"))
model.add(tf.keras.layers.MaxPooling2D(pool_size=(2,2)))

#third Convolutionnal layer
model.add(tf.keras.layers.Conv2D(64,(3,3),input_shape = (28,28)))
model.add(tf.keras.layers.Activation("relu"))
model.add(tf.keras.layers.MaxPooling2D(pool_size=(2,2)))

#tfully Connected layer layer
model.add(tf.keras.layers.Flatten())
model.add(tf.keras.layers.Dense(64))
model.add(tf.keras.layers.Activation("relu"))

model.add(tf.keras.layers.Dense(32))
model.add(tf.keras.layers.Activation("relu"))

model.add(tf.keras.layers.Dense(10))
model.add(tf.keras.layers.Activation("softmax"))
```

```python
#Show the summary of our model
model.summary()
```

```python
#Compiling the model
model.compile(optimizer='adam', loss= 'sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
#fitting the model
model.fit(x_trainr,y_train , epochs = 5, validation_split = 0.3)
```

```python
#Evaluating the model and printing its loss and accuracy on the test data
test_loss, test_acc = model.evaluate(x_testr, y_test)
print("Test loss on 10,000 test samples",test_loss)
print("Validation Accuracy on 10,000 test samples",test_acc)
```

```python
#predict the the values of the test data
predictions = model.predict([x_testr])
```

```python
### In this loop I will try to predict the images in my digits folder with my hand written model that I've saved earlier
##### with importing the file and accessing each image of it named digit and the transforme it into a gray picture
####### and then resize it and normalize it so we can pass it as an input to the model and predict its value.
image_number = 0
while os.path.isfile(f"digits/digit{image_number}.png"):
    try:
        img = cv2.imread(f"digits/digit{image_number}.png")
        img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
        resised = cv2.resize(img_gray, (28,28), interpolation = cv2.INTER_AREA)
        new = tf.keras.utils.normalize(resised, axis = 1)
        new = np.array(new).reshape(-1, img_size, img_size, 1)
        
        predictions = model.predict(new)
        print(f"This digit is probably a {np.argmax(predictions)}")
        plt.imshow(img)
        plt.show()
    except:
        print("Error!!!!")
    finally:
        image_number += 1
```
